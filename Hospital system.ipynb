```python
class Patient:
  def __init__(self, id, name, age, bloodgroup, next, prev):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev

class WRM:
  def __init__(self):
    self.head = Patient(None,None,None,None,None,None)
    self.last = self.head
  def RegisterPatient(self, id, name, age, bloodgroup):
    new_p = Patient(id,name,age,bloodgroup,None,self.last)
    self.last.next = new_p
    new_p.next = self.head
    self.head.prev = new_p
    self.last = new_p
    print("Added a patient successfully.")
  def ServePatient(self):
    node = self.head.next
    self.head.next = node.next
    node.next.prev = self.head
    return node.name
  def CancelAll(self):
    if(self.head.next == None):
      return False
    else:
      self.head.next = None
      self.head.prev = None
      self.last = self.head
      return True
  def CanDoctorGoHome(self):
    if(self.head.next == None and self.head.prev == None):
      print("Yes")
    else:
      print("No")
  def ShowAllPatient(self):
    if(self.head.next == None and self.head.prev == None):
      print("No Patient registered.")
    else:
      h = self.head
      h1 = self.head.next
      while(h != h1):
        print(h1.name, end = ' ')
        h1 = h1.next
      print()
  def ReverseTheLine(self):
    if(self.head.next == None and self.head.prev == None):
      return "No Patient"
    h = self.head
    h1 = self.head.prev
    new_head = Patient(None,None,None,None,None,None)
    while(h != h1):
      if(self.head.prev == h1):
        new = Patient(h1.id,h1.name,h1.age,h1.bloodgroup,None,new_head)
        new_head.next = new
      else:
        n1 = Patient(h1.id,h1.name,h1.age,h1.bloodgroup,None,new)
        new.next = n1
        new = n1
      h1 = h1.prev
    new.next = new_head
    new_head.prev = new
    self.last = new
    self.head = new_head
    print("successful")
hospital = WRM()
print('Welcome to Waiting Room Management System')
while True:
  print(f'==Choose an Option==\n1. RegisterPatient()\n2. ServePatient()\n3. CancelAll()\n4. CanDoctorGoHome()\n5. ShowAllPatient()\n6. ReverseTheLine()\n7. Exit\n====================')
  user = int(input('Enter your choice:'))
  if 1 > user or user > 7:
    print('Invalid number')
  elif user == 1:
    hospital.RegisterPatient(int(input('Enter Your ID:')),input('Enter Your Name:'),int(input('Enter Your Age:')),input('Enter Your bloodgroup:'))
    print('Success registering patient')
  elif user == 2:
    print(hospital.ServePatient())
  elif user == 3:
    if hospital.CancelAll() == True:
      print('Cancle All Patient')
    else:
      print('Unseccessful')
  elif user == 4:
    hospital.CanDoctorGoHome()
  elif user == 5:
    hospital.ShowAllPatient()
  elif user == 6:
    if hospital.ReverseTheLine() == None:
      pass
    else:
      print(hospital.ReverseTheLine())
  elif user == 7:
    print('Thank You for Using waiting Room management System\nEXITING....')
    break
```
